```python
# Imports class written to import shapefile and add exterior points:
from ImportShapefile import *
```

```python
# Update the link to where you have stored the shapefiles:
link = 'data/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp'

df_full = ImportShapefile(link).get_df()
```

## Filter on countries in Africa:

```python
mask = df_full['continent']=='Africa'

df = df_full.loc[mask,:].reset_index(drop=True)
```

## Creating 'area' column:
Adding a value we can plot with respect to (for demonstration)...

**Note:** I don't know the unit of the resulting areas. Numbers are a bit off
From what I can tell they're at least proportionate to the countries area...

```python
df['area'] = df['geometry'].area

#data_world['area'] = data_world['geometry'].area
```

## Drop all except a few columns:

```python
df = df.loc[:, ('name','subregion','x', 'y','area', 'pop_est')]
```
